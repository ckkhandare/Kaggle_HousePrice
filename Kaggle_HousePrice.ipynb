# Handle NA 

```python
df['LotFrontage']=df['LotFrontage'].fillna(0)
```

```python
df['Alley']=df['Alley'].fillna('No_Alley')
```

```python
df['MasVnrType']=df['MasVnrType'].fillna('None')
df['MasVnrArea']=df['MasVnrArea'].fillna(0)
```

```python
df['BsmtQual']=df['BsmtQual'].fillna('No_Basement')
df['BsmtCond']=df['BsmtCond'].fillna('No_Basement')
df['BsmtExposure']=df['BsmtExposure'].fillna('No_Basement')
df['BsmtFinType1']=df['BsmtFinType1'].fillna('No_Basement')
df['BsmtFinType2']=df['BsmtFinType2'].fillna('No_Basement')
```

```python
df['Electrical']=df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
df['FireplaceQu']=df['FireplaceQu'].fillna('No_fireplace')
```

```python
df['GarageType']=df['GarageType'].fillna('No_Garage')
df['GarageFinish']=df['GarageFinish'].fillna('No_Garage')
df['GarageCond']=df['GarageCond'].fillna('No_Garage')
df['GarageQual']=df['GarageQual'].fillna('No_Garage')
df['GarageYrBlt']=df['GarageYrBlt'].fillna(0)
```

```python
df['Fence']=df['Fence'].fillna('No_Fence')
df['PoolQC']=df['PoolQC'].fillna('No_PoolQC')
df['MiscFeature']=df['MiscFeature'].fillna('NONE')
```

# Feature selection

```python
select = SelectFromModel(RandomForestClassifier(n_estimators=100))
```

```python
select.fit(X,y)
```

```python
select.transform(X)
```

```python
selected_col= pd.DataFrame({'columns' : X.columns , 'imp_fet': list(select.get_support())})
```

```python
sorted_sel=selected_col.sort_values(by='imp_fet',ascending=False)
```

```python
sorted_sel
```

```python
sorted_sel= sorted_sel.loc[sorted_sel['imp_fet']==True]
```

```python
sorted_sel.shape
```

```python
sel_X= pd.DataFrame(X,columns = sorted_sel['columns'])
```

```python
sel_X.head()
```

# Train test split

```python
X_train,X_test,y_train,y_test=train_test_split(sel_X,y,test_size=0.10,random_state=7)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# model

```python
XGmodel = GradientBoostingRegressor(loss='ls',learning_rate=0.1,n_estimators=100,max_depth=20,random_state=7)
```

```python
XGmodel.fit(sel_X,y)
```

```python
y_predict=XGmodel.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predict)
rmse = sqrt(mse)
rmse
```
